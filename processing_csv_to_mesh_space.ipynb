thinking process is, we can algebraically create numbers to multiply all the lats and lngs by that will normalize the ones within the range, and anything above 0 and less than 1 is in our space,

```python
bb = [-113.0005555556, 31.99944444444, -111.9994444445, 33.00055555556]

lngs = {'min':bb[0],'max':bb[2]}
lats = {'min':bb[1],'max':bb[3]}
```

```python
(lats,lngs)
```

```python
df["norm_lat"] = (df["Latitude"] - lats['min'])/(lats['max'] - lats['min'])
```

```python
df["norm_lng"] = (df["Longitude"] - lngs['min'])/(lngs['max'] - lngs['min'])
```

using the &'s helps us increase our selectivity

```python
filtered = df[(df['norm_lat']> 0 )&(df['norm_lat']<1) & (df['norm_lng']> 0) & (df['norm_lng']<1)]
```

```python
filtered.to_csv("death_points_n33_w113.csv")
```
